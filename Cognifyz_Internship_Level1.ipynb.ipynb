## Geospatial Analysis

### Step 1: Installing and Importing Folium

Folium is a Python library used to create interactive maps. It helps visualize restaurant locations using latitude and longitude coordinates.

```python
!pip install folium
```

### Step 2: Importing Folium Library

The Folium library is imported to generate interactive geographical maps.

```python
import folium
```

### Step 3: Creating Base Map

A base map is created using the average latitude and longitude values from the dataset.

```python
restaurant_map = folium.Map(
    location=[df['Latitude'].mean(), df['Longitude'].mean()],
    zoom_start=5
)

restaurant_map
```

### Step 4: Adding Restaurant Locations

Markers are added to the map to display restaurant locations based on latitude and longitude values.

```python
for index, row in df.head(100).iterrows():
    folium.Marker(
        [row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name']
    ).add_to(restaurant_map)

restaurant_map
```

### Step 5: Restaurant Distribution by City

This analysis visualizes the number of restaurants available in different cities.

```python
import matplotlib.pyplot as plt
import seaborn as sns

city_counts = df['City'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=city_counts.index,
    y=city_counts.values
)

plt.xticks(rotation=45)

plt.title('Top 10 Cities by Restaurant Count')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')

plt.show()
```

### Step 6: Country Distribution Analysis

This step identifies countries with the highest number of restaurants in the dataset.

```python
country_counts = df['Country Code'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=country_counts.index,
    y=country_counts.values
)

plt.title('Restaurant Distribution by Country Code')
plt.xlabel('Country Code')
plt.ylabel('Restaurant Count')

plt.show()
```

### Step 7: Correlation Between Location and Rating

Correlation analysis is performed to determine whether restaurant location has any relationship with aggregate rating.

```python
correlation = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()

print(correlation)
```

### Step 8: Correlation Heatmap

A heatmap is created to visually represent correlations between latitude, longitude, and aggregate rating.

```python
plt.figure(figsize=(6,4))

sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```

### Step 9: Scatter Plot of Location and Ratings

A scatter plot is used to analyze how restaurant ratings vary with geographical locations.

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x=df['Longitude'],
    y=df['Latitude'],
    hue=df['Aggregate rating']
)

plt.title('Restaurant Locations and Ratings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```
